```python
import matplotlib.pyplot as p
import numpy as np
```

```python
#Accuracy Score = [XGBoost,RF,GBclf,LightGBM,DT,AdaBooster,CatBoost]
accuracy=[86.09,86.75,74.26,91.3,80.13,85.76,88.74]
```

```python
#Precision Score = [XGBoost,RF,GBclf,LightGBM,DT,AdaBooster,CatBoost]
precision=[83.44,86.09,88.41,82.11,83.33,83.77,90.75]
```

```python
#Recall Score = [XGBoost,RF,GBclf,LightGBM,DT,AdaBooster,CatBoost]
recall=[89.68,86.92,81.83,91.93,86.66,86.95,87.59]
```

```python
X=["XGBoost","RF","GBclf","LightGBM","DT","AdaBooster","CatBoost"]
x_axis=np.arange(len(X))
p.bar(x_axis-0.25,accuracy,0.25,label="accuracy")
p.bar(x_axis+0,precision,0.25,label="precision")
p.bar(x_axis+0.25,recall,0.25,label="recall")
p.xticks(x_axis, X)
p.xlabel("Classifiers")
p.ylabel("Score")
p.title("Evaluation score for each classifier")
p.legend()
p.show()
```
